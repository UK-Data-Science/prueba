```python
import pandas as pd

linkGit="https://github.com/UK-Data-Science/tarea2Sebastian/raw/main/dataUN%20-%20HDR21-22_Statistical_Annex_HDI_Table.csv"

data=pd.read_csv(linkGit)
data
```

```python
#CLEANING
#Columns
#Delete the columns that are not going to be used
byeColumns=['Back','Unnamed: 3','Unnamed: 5','Unnamed: 7','Unnamed: 9','Unnamed: 11','Unnamed: 13']
data.drop(columns=byeColumns,inplace=True)
```

```python
#Check the first 6 rows for data
data.head(6)
```

```python
#Replace the columns names with the third row values and their year 
data.columns=data.iloc[3,:]+" "+data.iloc[5,:]
data.columns
```

```python
#Rename the first column name
newColumns={data.columns[0]:data.iloc[4,0]}
data.rename(columns=newColumns,inplace=True)
data.columns
```

```python
#Replace spaces and weird symbols in the columns names
newColumns=data.columns.str.strip().\
                            str.replace(pat=r'\s',
                                         repl='_',
                                         regex=True).\
                              str.replace(pat=r'\W',
                                          repl='',
                                          regex=True)

newColumns={data.columns[0]:newColumns[0],data.columns[1]:newColumns[1],data.columns[2]:newColumns[2],data.columns[3]:newColumns[3],data.columns[4]:newColumns[4],data.columns[5]:newColumns[5],data.columns[6]:newColumns[6],data.columns[7]:newColumns[7]}
data.rename(columns=newColumns,inplace=True)
data.columns
```

```python
#Columns names too large, rename with a shorter name
newColumns={data.columns[1]:'HDI_2021',data.columns[2]:'Life_expec_2021',data.columns[3]:'Expec_years_school_2021',data.columns[4]:'Mean_years_school_2021',data.columns[5]:'GNI_2021',data.columns[6]:'GNI_minus_HDI_rank_2021'}
data.rename(columns=newColumns,inplace=True)
data.columns
```

```python
#Categories
#Search for any row used to classify the data
cat=data[pd.isnull(data.HDI_2021)]
cat.Country=cat.Country.str.replace(pat=r'\s',
                    repl='_',
                    regex=True)
cat.head(18)
```

```python
#Define the categories and their limits
data.loc[7:72,'Human_Dev']=cat.Country[6]
data.loc[74:122,'Human_Dev']=cat.Country[73]
data.loc[124:167,'Human_Dev']=cat.Country[123]
data.loc[169:200,'Human_Dev']=cat.Country[168]
data.loc[202:206,'Human_Dev']=cat.Country[201]
data.loc[208:214,'Human_Dev']=cat.Country[207]
data.loc[216:229,'Human_Dev']=cat.Country[215]

mapper1={cat.Country[6]:1,cat.Country[73]:2,cat.Country[123]:3,cat.Country[168]:4,cat.Country[201]:5.1,cat.Country[207]:5.2,cat.Country[215]:5.3}
data.Human_Dev.replace(mapper1,inplace=True)
```

```python
#Rows
#Check the first 8 rows
data.head(8)
```

```python
#Drop the first 7 rows
byeRows=range(0,7)
data.drop(index=byeRows,inplace=True)
```

```python
#Drop every row with a 'Nan' value in the Life_expec_2021 column
data[pd.isnull(data.Life_expec_2021)].index
data.drop(data[pd.isnull(data.Life_expec_2021)].index,inplace=True)
```

```python
#Rename the rows
data.index=range(0,210)
data
```

```python
#Data
#Remove possibles spaces in the data
def noSpaces(x,errors='coerce'):
  x=x.str.strip()
  return x

data.iloc[:,range(0,8)]=data.iloc[:,range(0,8)].apply(noSpaces)
```

```python
#Symbols
#Check symbols other than a dot ('.')
a=set(data.HDI_2021.str.extractall('(\W)').loc[:,0].to_list())-set('.')
b=set(data.Life_expec_2021.str.extractall('(\W)').loc[:,0].to_list())-set('.')
c=set(data.Expec_years_school_2021.str.extractall('(\W)').loc[:,0].to_list())-set('.')
d=set(data.Mean_years_school_2021.str.extractall('(\W)').loc[:,0].to_list())-set('.')
e=set(data.GNI_2021.str.extractall('(\W)').loc[:,0].to_list())-set('.')
f=set(data.GNI_minus_HDI_rank_2021.str.extractall('(\W)').loc[:,0].to_list())-set('.')
g=set(data.HDI_rank_2020.str.extractall('(\W)').loc[:,0].to_list())-set('.')
a,b,c,d,e,f,g
```

```python
#Replace or extract symbols
data.GNI_2021=data.GNI_2021.str.replace(',','') 
data.GNI_minus_HDI_rank_2021=data.GNI_minus_HDI_rank_2021.str.extract(pat=r'(^-\d+$|\d+)')
data.HDI_rank_2020=data.HDI_rank_2020.str.extract(pat=r'(^\w+)')
```

```python
#Check for other weird symbols in the HDI_2021 column
data[data.HDI_2021.str.contains(pat=r'\W')]
```

```python
#Keep the numerical data
data.HDI_2021=data.HDI_2021.str.extract(pat='(\d+)')
data.Expec_years_school_2021=data.Expec_years_school_2021.str.extract(pat='(\d+.\d+)')
data.Mean_years_school_2021=data.Mean_years_school_2021.str.extract(pat='(\d+.\d+)')
data.GNI_2021=data.GNI_2021.str.extract(pat='(\d+)')
```
